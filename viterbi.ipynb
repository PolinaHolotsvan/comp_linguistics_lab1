```python
tagged_sentences = brown.tagged_sents(tagset='universal')
tagged_sentences
```

```python
tagged_sentences = [[(word.lower(), tag) for word, tag in sentence] for sentence in tagged_sentences]
tagged_sentences
```

```python
sentences = []
tags = []
for tagged_sentence in tagged_sentences:
    sentence, tag = zip(*tagged_sentence)
    sentences.append(list(sentence))
    tags.append(list(tag))
```

```python
def get_initial_tag_probs(tags):
    initial_tag_counts = Counter([tag[0] for tag in tags])
    total_sentences = len(tags)
    initial_probs = {tag: count / total_sentences for tag, count in initial_tag_counts.items()}
    return initial_probs


initial_probs = get_initial_tag_probs(tags)
initial_probs
```

```python
tag_names = list(initial_probs.keys())
tag_names
```

```python
def get_transition_probs(tags):
    transition_counts = defaultdict(Counter)
    for tag_seq in tags:
        for i in range(1, len(tag_seq)):
            transition_counts[tag_seq[i - 1]][tag_seq[i]] += 1

    transition_probs = {tag: {next_tag: count / sum(transition_counts[tag].values())
                              for next_tag, count in next_tags.items()}
                        for tag, next_tags in transition_counts.items()}
    return transition_probs


transition_probs = get_transition_probs(tags)
transition_probs
```

```python
def get_emission_probs(sentences, tags):
    emission_counts = defaultdict(Counter)
    for sent, tag_seq in zip(sentences, tags):
        for word, tag in zip(sent, tag_seq):
            emission_counts[tag][word.lower()] += 1

    emission_probs = {tag: {word: count / sum(emission_counts[tag].values())
                            for word, count in words.items()}
                      for tag, words in emission_counts.items()}
    return emission_probs


emission_probs = get_emission_probs(sentences, tags)
emission_probs
```

```python
tag_num = len(tag_names)
tags_matrix = np.zeros((tag_num, tag_num), dtype='float32')
for i, tag1 in enumerate(tag_names):
    for j, tag2 in enumerate(tag_names):
        transition_prob = transition_probs[tag1][tag2]
        tags_matrix[i, j] = transition_prob

tags_matrix
```

```python
tags_df = pd.DataFrame(tags_matrix, columns=tag_names, index=tag_names)
tags_df
```

```python
def Viterbi(sentence_unsplit):
    states = []
    sentence = sentence_unsplit.split()

    for key, word in enumerate(sentence):
        p = []
        tr_p = []

        for tag in tag_names:
            if key == 0:  #first word has key=0 -> as the beginning starts after the end of the prev sentence 
                transition_prob = tags_df.loc['.', tag]
            else:
                transition_prob = tags_df.loc[states[-1], tag]

            state_probability = transition_prob
            emission_prob = 0

            if word in emission_probs[tag].keys():
                emission_prob = emission_probs[tag][word]

            state_probability *= emission_prob

            p.append(state_probability)
            tr_p.append(transition_prob)

        max_prob = max(p) if len(p) > 0 else max(tr_p)
        estimated_state = tag_names[p.index(max_prob)]
        states.append(estimated_state)
    return list(zip(sentence, states))
```

```python
def clear_sentence(sentence):
    punc = r'!()-[]{};:\'",<>./?@#$%^&*_~'
    for character in punc:
        sentence = sentence.replace(character, ' ' + character)
    return sentence
```
